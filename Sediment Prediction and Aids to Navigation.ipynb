Given Values

```
C = float(input('Bed load concentration(tons/day):' ))
Q = float(input('Water discharge rate(cubic feet per second):' ))
B = float(input('Width of river channel(m):' ))
H = float(input('Depth of river channel(m):' ))
p = float(input('Specific gravity of sediment:' ))
K = float(input('Value of K:'))
#Location
LatG = float(input('Latitude:' ))
LongG = float(input('Longitude:' ))
```

Unit Conversion

```
C = C*0.01049 
Q = Q*0.0283
p = p*1000
```

Sediment Level Prediction


```
Speed_of_water = Q/(B*H)
#Let speed of sediment be K times speed of water.
Speed_of_sediment = K*Q/(B*H)
L = (C/p)/(B*Speed_of_sediment)
Draught_limit = (H-L)
print('Limit of vessel Draught in metre :' )
print(Draught_limit)
```

Aids to navigation

Getting Location and Draught data

```
LatV = float(input('Latitude:' ))
LongV = float(input('Longitude:' ))
Draught_of_vessel = float(input('Draught of the vessel in metre:'))
```

Warning Generation

```
if ((LatV > LatG-2.5) and (LatV < LatG+2.5)): 
  if (Draught_of_vessel + 5) < Draught_limit :
    print('Safe')
  else: 
    print('Danger')
else: print('Vessel not in range')    
```
